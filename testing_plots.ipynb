```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib as mpl
```

```python
fig, (ax1, ax2) = plt.subplots(2,1,gridspec_kw = {'height_ratios':[3, 1]})
fig.set_figheight(12)
fig.set_figwidth(9)
fig.tight_layout()
fig.subplots_adjust(hspace=0.01)

a = np.random.rand(100).reshape(10,10)

im = ax1.imshow(a)
cbar = plt.colorbar(mappable=im, ax = ax1, fraction=0.047, pad=0.01)

x = np.arange(100)
ax2.plot(x, a.ravel(), label='some graph')
ax2.legend(loc=2)
ax2.set_ylabel('random')
ax2.set_xlabel('length [a.u]')
```
